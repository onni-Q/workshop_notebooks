## Real-life Dataset Exercise

As a final exercise, we use our knowledge to analyse a famous real-life social dataset about death penalties. We will work through a detailed example of data exploration and analyse a very small part of the complex question of racism in the US legal system.

The data `deathpenalty_array` covers judgments of defendants in cases of multiple murders in Florida between 1976 and 1987. The cases all have features that (a) describe whether the death penalty was handed out (where 0 refers to no, 1 to yes), (b) the race of the victim and (c) the race of the defendant (black is referred as 0, white is 1).

It is very problematic to record non-binary attributes like 'race' in this way. 'race' is a highly contested idea, and we should be more careful than the data we you use, which assigns people to two very simplified categories. Discussing such constraints should be part of your own critical data analysis. But you are only at the beginning of your programming journey, and we will ignore this for now to make things not too complicated.

Load the dataset by running the cell below. Take a moment to think about what you have just loaded. How is the data organized?

````python
#Keep cell

import numpy as np

deathpenalty_array = np.array([
[0,	0,	0,	139],
[1,	0,	0,	4],
[0,	1,	0,	37],
[1,	1,	0,	11],
[0,	0,	1,	16],
[1,	0,	1,	0],
[0,	1,	1,	414],
[1,	1,	1,	53]])

deathpenalty = deathpenalty_array[:,0]
victimrace = deathpenalty_array[:,1]
defendantrace = deathpenalty_array[:,2]
freq = deathpenalty_array[:,3]

print('Death Penalty :', deathpenalty)
print('Victim Race   :', victimrace)
print('Defendant Race:', defendantrace)
print('Frequency     :', freq)
````

Remember that a black person is represented by 0 and a white by 1. A death penalty was handed out if the column `deathpenalty` equals 1, with 0 otherwise. This is quite a confusing way of recording events and not recommended beyond the use in such a small example. It was, however, quite common in those early days of data.

`deathpenalty_array` records all the information, and we use indexes to select particular columns of the data to define:

- The array `freq` for frequency of each combination, which has 674 cases in total.
- `deathpenalty`: Was the judgment death penalty? yes = 1, no = 0.
- `victimrace`: the race of the victim; white = 1, black = 0.
- `defendantrace`: the race of the defendant; white = 1, black = 0.
