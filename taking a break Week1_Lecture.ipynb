We want to print out the 'The lists have it!' each time the `friends_list` element at position `i` is larger or equal the `friends_array` element at position `i`. Otherwise, we print 'The arrays have it!'.

Take a moment to think how to do it. You need both for-loops and if-statements.

After some thinking, you type:

```
for i in list(range(5)):
    if (friends_list[i]>=friends_array[i]):
        print('The lists have it!')
    else:
        print('The arrays have it!')
```
